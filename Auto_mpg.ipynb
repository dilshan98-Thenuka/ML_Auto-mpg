```python
data = pd.read_fwf("/content/drive/MyDrive/SCS3201_ML/Auto-mpg/auto-mpg.data",
                   names=["mpg","cylinders","displacement","horsepower","weight","acceleration","model_year","origin","car_name"])
data
```

```python
data['horsepower'] = pd.to_numeric(data['horsepower'], errors='coerce')
```

```python
import math
median_horsepwr = math.floor(data.horsepower.median())
data.horsepower = data.horsepower.fillna(median_horsepwr)
data
```

```python
data.dtypes
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    data[['cylinders',	'displacement',	'horsepower',	'weight',	'acceleration']],data.mpg,train_size=0.9,random_state=42
    )
```

```python
from sklearn.linear_model import LinearRegression
mlReg = LinearRegression()
mlReg.fit(X_train,y_train)
```

```python
mlReg.coef_
```

```python
mlReg.intercept_
```

```python
mlReg.predict(X_test)
```

```python
mlReg.score(X_test,y_test)
```
